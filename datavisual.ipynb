```
import matplotlib.pyplot as plt
x=[2,5,4,9]
y=[3,8,3,6]
```

```
plt.xlabel("distance")#for laveling x axis 
plt.ylabel("time") #for y
plt.scatter(x,y,label='minss')# to show dots only
plt.plot(x,y,label='raiway track') # to plot line
plt.bar(x,y,label='finace growth')# to show bar plot
plt.bar(y,x,color='yellow')
plt.grid(c='green')#for grid and colour as well
plt.legend()
plt.plot() # to plot the graph
```

```
players=['dhoni','rohit','virat']
runs=[120,160,130]
```

```
plt.xlabel('players')
plt.ylabel('runs')
plt.bar(players,runs,label='2020 starting match')
plt.legend()
plt.show()
```

```
plt.xlabel('players')
plt.ylabel('runs')
plt.pie(runs,labels=players,autopct='%1.1f%%',explode=(0,0.3,0),shadow=True)
plt.legend()
plt.show()
```
